```python
import itertools
import pandas as pd

# Define the possible numbers
numbers = [0, 1, 2, 3, 4]

# Generate all possible combinations of the numbers for 8 people
combinations = itertools.product(numbers, repeat=8)

# Convert the combinations to a list of tuples
combo_list = list(combinations)

# Print the first 10 tuples in the list for verification
# print(combo_list[:10])

combo_list = pd.DataFrame(combo_list)
```

```python
combo_list['sum'] = combo_list[0]+combo_list[1]+combo_list[2]+combo_list[3]+combo_list[4]+combo_list[5]+combo_list[6]+combo_list[7]
combo_list = combo_list[combo_list['sum']==20]
combo_list
```
