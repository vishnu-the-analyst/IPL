```python
import glob
import numpy as np
csv_files = (glob.glob("C:/Users/Vishnu Kumar/Documents/IPL/raw/*info.csv")) 
```

```python
import pandas as pd


df_append = pd.DataFrame()

for file in csv_files:
            df_temp = pd.read_csv(file,sep='漢ழ்字', engine='python',header=None)
            df_temp['ID'] = file
            df_temp = df_temp[(df_temp[0].str.startswith('info,toss_winner')) | (df_temp[0].str.startswith('info,toss_decision'))]
            df_append = pd.concat([df_append,df_temp])

df_append = df_append.reset_index()
df_append[['info', 'attrib','value']] = df_append[0].str.split(",", expand = True)
df_append = df_append.drop(columns=['index',0,'info'])
df_append['ID'] = df_append['ID'].str[40:-9]
toss_winner = df_append[df_append['attrib']=='toss_winner'].copy()
toss_decision = df_append[df_append['attrib']=='toss_decision'].copy()
toss_winner['ID'] = toss_winner['ID'].astype('int')
toss_decision['ID'] = toss_decision['ID'].astype('int')
```

```python
all_matches = pd.read_csv("C:/Users/Vishnu Kumar/Documents/IPL/raw/all_matches.csv")
# all_matches
all_matches = pd.merge(all_matches, toss_winner, how="left" ,left_on = "match_id", right_on = "ID" )
all_matches = all_matches.drop(columns=["ID", "attrib"])
all_matches = all_matches.rename(columns={"value": "toss_winner"})
all_matches = pd.merge(all_matches, toss_decision, how="left" ,left_on = "match_id", right_on = "ID" )
all_matches = all_matches.drop(columns=["ID", "attrib"])
all_matches = all_matches.rename(columns={"value": "toss_decision"})
all_matches = all_matches[all_matches['innings']<=2]
all_matches['season'] = all_matches['season'].replace(['2007/08', '2009', '2009/10', '2011' ,'2012','2018', '2019', '2020/21', '2021'], [2008,2009,2010,2011,2012,2018,2019,2020,2021])
all_matches['toss_win_batting'] = all_matches.apply(lambda x: 1 if x['toss_winner'] == x['batting_team'] else 0, axis=1)
all_matches['toss_decision'] = all_matches.apply(lambda x: 1 if x['toss_decision'] == 'bat' else 0, axis=1)
all_matches['player_dismissed'] = all_matches.apply(lambda x: 1 if x['player_dismissed'] == x['striker'] else 0, axis=1)
all_data = all_matches.copy()
# all_matches = pd.concat([all_matches, toss_winner], axis=1, join='outer', join_axes=[match_id,ID])
```

```python
all_for_bowlers = all_data.copy()
all_for_bowlers['ball'] = (all_for_bowlers['ball'].apply(np.floor))+1
all_for_bowlers = all_for_bowlers.drop(columns=['start_date','striker','non_striker','runs_off_bat','extras','wides','noballs','byes','legbyes','penalty','wicket_type','player_dismissed','other_wicket_type','other_player_dismissed'])
all_for_bowlers['over_1'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 1 else 0, axis=1)
all_for_bowlers['over_2'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 2 else 0, axis=1)
all_for_bowlers['over_3'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 3 else 0, axis=1)
all_for_bowlers['over_4'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 4 else 0, axis=1)
all_for_bowlers['over_5'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 5 else 0, axis=1)
all_for_bowlers['over_6'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 6 else 0, axis=1)
all_for_bowlers['over_7'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 7 else 0, axis=1)
all_for_bowlers['over_8'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 8 else 0, axis=1)
all_for_bowlers['over_9'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 9 else 0, axis=1)
all_for_bowlers['over_10'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 10 else 0, axis=1)
all_for_bowlers['over_11'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 11 else 0, axis=1)
all_for_bowlers['over_12'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 12 else 0, axis=1)
all_for_bowlers['over_13'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 13 else 0, axis=1)
all_for_bowlers['over_14'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 14 else 0, axis=1)
all_for_bowlers['over_15'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 15 else 0, axis=1)
all_for_bowlers['over_16'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 16 else 0, axis=1)
all_for_bowlers['over_17'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 17 else 0, axis=1)
all_for_bowlers['over_18'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 18 else 0, axis=1)
all_for_bowlers['over_19'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 19 else 0, axis=1)
all_for_bowlers['over_20'] = all_for_bowlers.apply(lambda x: 1 if x['ball'] == 20 else 0, axis=1)
# all_for_bowlers
```

```python
all_for_bowlers = pd.pivot_table(all_for_bowlers, values =['over_1','over_2','over_3','over_4','over_5','over_6','over_7','over_8','over_9','over_10','over_11','over_12','over_13','over_14','over_15','over_16','over_17','over_18','over_19','over_20'], index =['match_id', 'season','venue','innings','batting_team','bowling_team','bowler','toss_winner','toss_decision','toss_win_batting'], aggfunc = np.max).reset_index()
all_for_bowlers = pd.pivot_table(all_for_bowlers, values=['over_1','over_2','over_3','over_4','over_5','over_6','over_7','over_8','over_9','over_10','over_11','over_12','over_13','over_14','over_15','over_16','over_17','over_18','over_19','over_20'], index=['match_id','venue','innings','bowler','batting_team','bowling_team'], aggfunc=np.sum).reset_index()
```

```python
all_for_bowlers = all_for_bowlers.rename(columns={"match_id":"match number", "bowler": "bowler name","bowling_team" : "team name","batting_team": "opposite team name","innings":"bowling innings" })
```

```python
import pandas as pd
import numpy as np

# load the dataset
dataset = all_for_bowlers.copy()

# drop irrelevant columns
dataset = dataset.drop(columns=['match number', 'venue'])

# one-hot encode categorical features
dataset = pd.get_dummies(dataset, columns=['bowler name', 'team name', 'opposite team name'])

# create a new feature for number of overs bowled by each bowler
# overs_cols = [f"over_{i}" for i in range(1, 21)]
# dataset['total_overs'] = dataset[overs_cols].sum(axis=1)

# create a binary feature for the bowling innings
dataset['bowling_innings'] = dataset['bowling innings'].apply(lambda x: 1 if x == 1 else 0)
dataset = dataset.drop(columns=['bowling innings'])

# # normalize the data
# cols_to_normalize = dataset.columns[3:]
# dataset[cols_to_normalize] = (dataset[cols_to_normalize]-dataset[cols_to_normalize].min())/(dataset[cols_to_normalize].max()-dataset[cols_to_normalize].min())

# split the dataset into train and test sets
train_set = dataset.sample(frac=0.8, random_state=42)
test_set = dataset.drop(train_set.index)

# extract the features and labels
X_train = train_set.iloc[:, 20:].values
y_train = train_set.iloc[:, 0:20].values
X_test = test_set.iloc[:, 20:].values
y_test = test_set.iloc[:, 0:20].values

```

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# create a neural network model
model = Sequential()

# add a dense layer with 128 units and relu activation
model.add(Dense(128, activation='relu', input_dim=X_train.shape[1]))

# add dropout regularization
model.add(Dropout(0.2))

# add another dense layer with 64 units and relu activation
model.add(Dense(64, activation='relu'))

# add output layer with sigmoid activation
model.add(Dense(1, activation='sigmoid'))  # output 20 probabilities for 20 overs

# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)

# predict on test data
y_pred = model.predict(X_test)

# evaluate the model
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Loss: {loss:.4f}, Accuracy: {accuracy*100:.2f}%")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy

# Load the dataset
df = all_for_bowlers.copy()

# Encode categorical features
le = LabelEncoder()
df['bowler name'] = le.fit_transform(df['bowler name'])
# df['team name'] = le.fit_transform(df['team name'])
df['opposite team name'] = le.fit_transform(df['opposite team name'])
df['venue'] = le.fit_transform(df['venue'])

# Split the dataset into training and testing sets
train_df = df.sample(frac=0.8, random_state=42)
test_df = df.drop(train_df.index)

# Create X and y arrays for training and testing
X_train = train_df.drop(columns=['venue','team name','match number', 'over_1', 'over_2', 'over_3', 'over_4', 'over_5', 'over_6', 'over_7', 'over_8', 'over_9', 'over_10', 'over_11', 'over_12', 'over_13', 'over_14', 'over_15', 'over_16', 'over_17', 'over_18', 'over_19', 'over_20'])
y_train = train_df[['over_1', 'over_2', 'over_3', 'over_4', 'over_5', 'over_6', 'over_7', 'over_8', 'over_9', 'over_10', 'over_11', 'over_12', 'over_13', 'over_14', 'over_15', 'over_16', 'over_17', 'over_18', 'over_19', 'over_20']].values

X_test = test_df.drop(columns=['venue','team name','match number', 'over_1', 'over_2', 'over_3', 'over_4', 'over_5', 'over_6', 'over_7', 'over_8', 'over_9', 'over_10', 'over_11', 'over_12', 'over_13', 'over_14', 'over_15', 'over_16', 'over_17', 'over_18', 'over_19', 'over_20'])
y_test = test_df[['over_1', 'over_2', 'over_3', 'over_4', 'over_5', 'over_6', 'over_7', 'over_8', 'over_9', 'over_10', 'over_11', 'over_12', 'over_13', 'over_14', 'over_15', 'over_16', 'over_17', 'over_18', 'over_19', 'over_20']].values

# Define the model architecture
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(32, activation='relu'))
model.add(Dense(20, activation='sigmoid'))

# Compile the model
optimizer = Adam(learning_rate=0.001)
loss_fn = BinaryCrossentropy()
model.compile(optimizer=optimizer, loss=loss_fn)

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

```python
# predict on test data
y_pred = model.predict(X_test)

# evaluate the model
accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Loss: {loss:.4f}, Accuracy: {accuracy*100:.2f}%")

```
