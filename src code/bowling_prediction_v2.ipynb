```python
import pandas as pd
import numpy as np

# load the dataset
dataset = all_for_bowlers.copy()

# drop irrelevant columns
dataset = dataset.drop(columns=['match number', 'venue','ID','team','opposite team name','team name'])

# one-hot encode categorical features
dataset = pd.get_dummies(dataset, columns=['bowler name'])

# create a new feature for number of overs bowled by each bowler
# overs_cols = [f"over_{i}" for i in range(1, 21)]
# dataset['total_overs'] = dataset[overs_cols].sum(axis=1)

# create a binary feature for the bowling innings
dataset['bowling_innings'] = dataset['bowling innings'].apply(lambda x: 1 if x == 1 else 0)
dataset = dataset.drop(columns=['bowling innings','bowling_innings'])

# # normalize the data
# cols_to_normalize = dataset.columns[3:]
# dataset[cols_to_normalize] = (dataset[cols_to_normalize]-dataset[cols_to_normalize].min())/(dataset[cols_to_normalize].max()-dataset[cols_to_normalize].min())

# split the dataset into train and test sets
train_set = dataset.sample(frac=0.8, random_state=42)
test_set = dataset.drop(train_set.index)

# extract the features and labels
X_train = train_set.iloc[:, 20:].values
y_train = train_set.iloc[:, 0:20].values
X_test = test_set.iloc[:, 20:].values
y_test = test_set.iloc[:, 0:20].values

```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy

# Define the model architecture
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(32, activation='relu'))
model.add(Dense(20, activation='sigmoid'))

# Compile the model
optimizer = Adam(learning_rate=0.001)
loss_fn = BinaryCrossentropy()
model.compile(optimizer=optimizer, loss=loss_fn)

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

```python
import sklearn.metrics as metrics

# predict on test data
y_pred = model.predict(X_test)

# evaluate the model
test_loss = model.evaluate(X_test, y_test, verbose=0)
accuracy = metrics.accuracy_score(y_test, y_pred.round())

print(f"Loss: {test_loss:.4f}, Accuracy: {accuracy*100:.2f}%")
```
